```python
from z3 import * #Optimize for boolean model 
import numpy as np #operate array
import matplotlib.pyplot as plt #show image
import matplotlib
import pandas as pd
import time #calculate method running time
%config InlineBackend.figure_format = 'retina' #make a retina image
```

```python
# Check our model are Sat ?
def Fitt_rect(FW,FH,width,height):
    
    Width = FW #Width side of Big rectangle
    Height = FH #Height side of Big rectangle
    W = width #list of width of small rectangle
    H = height #list of height of small rectangle
    n = len(W)
    
    #define variable of model 
    w = [Int(f'w{i}') for i in range(n)] 
    h = [Int(f'h{i}') for i in range(n)]
    x = [Int(f'x{i}') for i in range(n)]
    y = [Int(f'y{i}') for i in range(n)]
    
    #define rotate a rectangle condition 
    Form = [Or(And(w[i]==W[i],h[i]==H[i]),
               And(w[i]==H[i],h[i]==W[i])) for i in range(n)]

    #define small rectangles are in a Big rectangle
    IN = [And(x[i]>=0,x[i]+w[i]<=Width,y[i]>=0,y[i]+h[i]<=Height) for i in range(n)]
    
    #define all of small rectangle are no-overlab
    Overlap = []
    for i,j in [(m,k) for m in range(n) for k in range(n) if m < k]:
        Overlap = Overlap + [Or(x[i]+w[i]<=x[j],x[j]+w[j]<=x[i],y[i]+h[i]<=y[j],y[j]+h[j]<=y[i])]
    
    #add 3 conditions in a solver
    s = Solver()
    s.add(Form + IN + Overlap)
    
    #check a model is sat or unsat
    if s.check() == sat:
        m = s.model()
        return [(m[x[i]],m[y[i]],m[w[i]],m[h[i]]) for i in range(n)], True
    else:
        return _, False
```

```python
#Find minimum the big square rectangle that can contained all of small rectangles.
def Search(W, H, width, height,step=1):
    FW = W
    FH = H
    while True:

        Model, Check = Fitt_rect(FW, FH,width,height)
        #check a model is sat?
        if Check:
            print('Model : ', Model)
            print(f'W_side : {FW}\nH_side : {FH}')
            print(f'Area : {FW*FH}')
            print('-------------------------------------------')
            return Model, FW, FH
            break
            
        else:
            #if a model is not sat, we will expand a big square rectangle.
            FW+=step
            FH+=step
```

```python
#Make a list of small rectangles.
RECT = [(1,2)]*21 + [(2,5)]*15 + [(15,6)]*7
# 2*20+10*10+25*6*6
size = [np.prod([i]) for i in RECT]
S = sum(size)
#Find a length of side of the big square rectangle.
side = int(np.sqrt(S))+1
#saperate width side and height side of small rectangles.
RECT_W = [i for i, _ in RECT]
RECT_H = [i for _, i in RECT]

#search a answer of model
s = time.time()
model, W, H = Search(side, side, RECT_W, RECT_H, step=1)
e = time.time()
#show running time
print('running time : ', e-s)

#show a rect image.
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
for i,m in enumerate(model):
    y, x, w, h = m
    
    #change z3 type to int type
    x = x.as_long()
    y = y.as_long()
    w = w.as_long()
    h = h.as_long()
    
    #draw a small rectangle in the big rectangle
    rect = matplotlib.patches.Rectangle((y,x), w, h, color='gray', linewidth = 1,ec ='black')

    ax.add_patch(rect)

plt.xlim([0, W])
plt.ylim([0, H])
plt.savefig('RECT.png')
plt.show()
```

```python
#Make a list of small rectangles.
RECT_list = [[(1,2)]*i + [(2,5)]*j + [(15,6)]*k for i, j, k in [(1,2,2), (5,2,3), (5,5,5), (7,12,1), (15,5,5), (20,5,5)]]
T = []
A = []

for index,RECT in enumerate(RECT_list):
    #Find a length of side of the big square rectangle.
    size = [np.prod([i]) for i in RECT]
    S = sum(size)
    side = int(np.sqrt(S))+1
    #saperate width side and height side of small rectangles.
    RECT_W = [i for i, _ in RECT]
    RECT_H = [i for _, i in RECT]

    s = time.time()
    model, W, H = Search(side, side, RECT_W, RECT_H, step=1)
    e = time.time()
    
    #show running time
    t = e-s
    print('running time : ',t)
    T.append(t)
    A.append(W*H)

    #show a rect image.
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    for i,m in enumerate(model):
        y, x, w, h = m
        #change z3 type to int type
        x = x.as_long()
        y = y.as_long()
        w = w.as_long()
        h = h.as_long()
        
        #draw a small rectangle in the big rectangle
        rect = matplotlib.patches.Rectangle((y,x), w, h, color='gray', linewidth = 1,ec ='black')

        ax.add_patch(rect)
    
    plt.xlim([0, W])
    plt.ylim([0, H])
    plt.savefig(f'RECT_{index}.png')
    plt.show()
```

```python
#show table
DF = pd.DataFrame({'TIME (S)' : T, 
                      'Number of parts' : [5,10,15,20,25,30], 
                      'AREA_use': A, 
                      'AREA_product': [202, 300, 510, 224, 530, 540]})
DF['LOST_AREA'] = DF['AREA_use'] - DF['AREA_product']
DF
```
